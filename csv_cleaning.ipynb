```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
platform_path = "resources/platform.csv"
publisher_path = "resources/publisher.csv"
```

```python
publisher = pd.read_csv(publisher_path, encoding = "ISO-8859-1")
publisher
```

```python
publisher_db = publisher.drop(['Notable games published', 'Notes', 'Active'], axis=1)
publisher_db
```

```python
publisher_db.rename(columns = {'Publisher':'publisher','Headquarters':'pub_hq', 'Est.':'pub_est'}, inplace = True)

# publisher_db.to_csv('output_data/publisher_clean.csv', encoding = 'utf-8', 
#                     index = False, header = True)
publisher_db
```

```python
platform = pd.read_csv(platform_path, encoding = "ISO-8859-1")
platform
```

```python
platform_db = platform.drop(['nb_jeu', 'sortie_eu', 'sortie_usa', 'gen', 'sortir_jp'], axis=1)
platform_db
```

```python
platform_db.rename(columns = {'Initiales':'platform', 'nom':'full_platform',
                           'manufactureur':'platform_manufacturer'}, inplace = True)

# platform_db.to_csv('output_data/platform_clean.csv', encoding = 'utf-8', 
#                     index = False, header = True)

platform_db
```

```python
#vg sales code begins- Rob

# read csv file
csv_file = "resources/vgsales.csv"
vg_df = pd.read_csv(csv_file)

# set index by rank
vg_df.set_index(['Rank'], inplace=True)
vg_df
```

```python
# drop blanks
vg_df.dropna(
    axis=0,
    how='any',
    inplace=True
)
vg_df
```

```python
# change data type of year to int in order to remove decimal place
vg_df['Year'] = vg_df['Year'].astype(int)
vg_df
```

```python
vg_df.rename(columns = {'Rank':'rank','Name':'name','Platform':'platform', 'Year':'year',
                       'Genre':'genre','Publisher':'publisher', 'NA_Sales':'na_sales',
                       'EU_Sales':'eu_sales', 'JP_Sales':'jp_sales', 'Other_Sales':'other_sales',
                       'Global_Sales':'global_sales'}, inplace = True)
vg_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/VG_Sales_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
publisher_db.to_sql(name='publisher', con=engine, if_exists='append', index=False)
```

```python
platform_db.to_sql(name='platform', con=engine, if_exists='append', index=False)
```

```python
vg_df.to_sql(name='vg_sales_clean', con=engine, if_exists='append', index=False)
```
