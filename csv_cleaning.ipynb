```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
platform_path = "resources/platform.csv"
developer_path = "resources/developer.csv"
publisher_path = "resources/publisher.csv"
```

```python
publisher = pd.read_csv(publisher_path, encoding = "ISO-8859-1")
publisher
```

```python
publisher_db = publisher.drop(['Notable games published', 'Notes', 'Active', 'Headquarters'], axis=1)
publisher_db
```

```python
publisher_db.rename(columns = {'Publisher':'publisher', 'Est.':'pub_est'}, inplace = True)

publisher_db.to_csv('output_data/publisher_clean.csv', encoding = 'utf-8', 
                    index = False, header = True)
publisher_db
```

```python
developer = pd.read_csv(developer_path, encoding = "ISO-8859-1")
developer
```

```python
developer_db = developer.drop(['Notable games, series or franchises', 'Notes', 'Active',
                               'Autonomous area','City'], axis=1)
developer_db
```

```python
developer_db.rename(columns = {'Developer':'developer', 'Country':'dev_country',
                              'Est.':'dev_est'}, inplace = True)

developer_db.to_csv('output_data/developer_clean.csv', encoding = 'utf-8', 
                    index = False, header = True)
developer_db
```

```python
platform = pd.read_csv(platform_path, encoding = "ISO-8859-1")
platform
```

```python
platform_db = platform.drop(['nb_jeu', 'sortie_eu', 'sortie_usa', 'gen', 'sortir_jp'], axis=1)
platform_db
```

```python
platform_db.rename(columns = {'Initiales':'platform', 'nom':'pform_full',
                           'manufactureur':'pform_manuf'}, inplace = True)

platform_db.to_csv('output_data/platform_clean.csv', encoding = 'utf-8', 
                    index = False, header = True)

platform_db
```

```python
#vg sales code begins- Rob

# read csv file
csv_file = "resources/vgsales.csv"
vg_df = pd.read_csv(csv_file)

# set index by rank
vg_df.set_index(['Rank'], inplace=True)
vg_df
```

```python
# drop blanks
vg_df.dropna(
    axis=0,
    how='any',
    inplace=True
)
vg_df
```

```python
# change data type of year to int in order to remove decimal place
vg_df['Year'] = vg_df['Year'].astype(int)
```

```python
vg_df
```

```python
# save csv to output folder
vg_df.to_csv('output_data/vg_sales_clean.csv')
```
